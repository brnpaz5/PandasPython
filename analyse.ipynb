```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt 
plt.style.use('seaborn')
```

```python
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
pd.options.display.float_format = '{:20,.2f}'.format #Ajustado exibição de Valores  
```

```python
df.groupby(df['Data Venda'].dt.year)['Valor Venda'].sum() #Valores de vendas por Ano
```

```python
df.groupby([df['Data Venda'].dt.year, 'Cor'])['Valor Venda'].sum() #Valor de Venda por Ano e Marca
```

```python
df['Lucro'] = df['Valor Venda'] - df['Custo Unitário']
```

```python
df.head(10)
```

```python
lucro_ano_cor = df.groupby([df['Data Venda'].dt.year, 'Cor'])['Lucro'].sum()
```

```python
lucro_ano_cor.plot.barh(title='Vendas por Cor') #Vendas por Cor e Ano
```

```python
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto')
```

```python
df_2008 = df.loc[df['Data Venda'].dt.year == 2008] #Capturando vendas de 2008
df_2009 = df.loc[df['Data Venda'].dt.year == 2009] #Capturando vendas de 2009
```

```python
df_2008.head(5)
```

```python
df_2008.groupby('Produto')['Lucro'].sum().plot.pie(title='Lucro por Produto 2008') #Grafico de Lucro por Produto de 2008
```

```python
df_2009.groupby('Produto')['Lucro'].sum().plot.pie(title='Lucro por Produto 2009') #Grafico de Lucro por Produto de 2009
```

```python
df_2009.groupby('Produto')['Lucro'].sum()
```

```python
df_2008.groupby(df['Data Venda'].dt.month)['Lucro'].sum().plot()
```
